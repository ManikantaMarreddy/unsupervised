```python
import pandas as pd
import numpy as np

X = np.random.uniform(0,1,50)
Y = np.random.uniform(0,1,50)
data_frame = pd.DataFrame(columns=["X","Y"])
data_frame.X = X
data_frame.Y = Y
data_frame

data_frame.plot(x = "X",y = "Y", kind= "scatter")
```
